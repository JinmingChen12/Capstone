```python
import requests
from bs4 import BeautifulSoup

with open("C:/Users/11946/NBA2018.html", "rb") as fp:
    soup = BeautifulSoup(fp, 'html.parser')
print(soup)
```

```python
info = []
all_players_div = soup.find('div', class_='nba-stat-table__overflow')
all_players_body = soup.find('tbody')
all_players_rows = all_players_body.find_all("tr")
for player_row in all_players_rows:
     all_cols = player_row.find_all("td")
     info.append({
         "Player": all_cols[0].get_text(),
         "POS": all_cols[1].get_text(),
         "Body Fat %": all_cols[2].get_text(),
         "Hand Length(Inches)": all_cols[3].get_text(),
         "Hand Width(Inches)":all_cols[4].get_text(),
         "Height W/O Shoes": all_cols[5].get_text(),
         "Height W/ Shoes": all_cols[6].get_text(),
         "Standing Reach": all_cols[7].get_text(),
         "Weight(lbs)": all_cols[8].get_text(),
         "Wingspan": all_cols[9].get_text(),         
     })
print(info)
```
